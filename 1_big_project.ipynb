#### Какие платформы лидируют по продажам, растут или падают? Выбрать несколько потенциально прибыльных платформ

__Какие платформы выбрать?__ Просмотрев таблицу, можно сделать вывод, что нужно взять платформы из Группы 1, Группы 2 и Группы 3 (т.е. 5 платформ).

Значит распределение по годам буду строить именно по этим трём группам, то есть по платформам значение суммы которых от 100 и выше. Что мне надо - свести всё вместе, где указать ежегодный объём продаж, отфильтровать значения от 2011 года и выше и, собственно всё вывести на экран.

```python
ps3 = games.query('platform == "ps3"').pivot_table(index='year_of_release', values='sales_grouped', aggfunc='sum')
x360 = games.query('platform == "x360"').pivot_table(index='year_of_release', values='sales_grouped', aggfunc='sum')
ps4 = games.query('platform == "ps4"').pivot_table(index='year_of_release', values='sales_grouped', aggfunc='sum')
ds3 = games.query('platform == "3ds"').pivot_table(index='year_of_release', values='sales_grouped', aggfunc='sum')
xone = games.query('platform == "xone"').pivot_table(index='year_of_release', values='sales_grouped', aggfunc='sum')

top_5 = (
    ps3.merge(x360, how='outer', on='year_of_release')
    .merge(ps4, how='outer', on='year_of_release')
    .merge(ds3, how='outer', on='year_of_release')
    .merge(xone, how='outer', on='year_of_release')
)

# Ещё переименую названия столбиков, чтобы было всё удобнее
top_5.set_axis(['ps3', 'x360', 'ps4', '3ds', 'xone'], axis='columns', inplace=True)
```

```python
top_5_2011 = top_5.query('year_of_release >= 2014')

top_5_2011.plot(stacked=True, figsize=(16, 9), title='Объём продаж с 2014 по 2016 по 5 платформам',
                grid=True).set(xlabel='Релизный год', ylabel='Объём продаж');
```

## Гипотезы

### средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Нужно сделать Series для платформ с оценками пользователей по нужному срезу лет и по нужным вопросам.

Буду проверять гипотезы scipy.stats.ttest_ind()

__Выбрать alpha__

Есть стандартные уровни значимости 0,1; 0,05; 0,01; 0,001.

Если объем выборки небольшой до 100 единиц, то можно вполне отвергнуть нулевую гипотезу при уровне значимости 0,05 или даже 0,1. При объеме выборки, измеряемой сотнями – от 100 до 1000, следует понизить уровень значимости хотя бы до 0,01. А при больших выборках, измеряемых тысячами наблюдений, уверенно отвергать нулевую гипотезу можно только при значимости меньшей 0,001. Соответственно нужно взять alpha = 0,05 для проверки гипотез. 

Соответственно, выбираю 0.05.

Два условия:

- H0 средние рейтинги юзеров Xbox One и PC равны

- H1 средние рейтинги юзеров Xbox One и PC не равны

Из стобца platform мне нужны только xone и pc, а также определённые выше года с 2006 по 2016. Так как мне нужно использовать рейтинги юзеров - user_score.

```python
for_xbx1 = games.query('platform == "xone" and year_of_release >= 2014 and user_score < 101 and critic_score < 101')['user_score']
for_pc = games.query('platform == "pc" and year_of_release >= 2014 and user_score < 101 and critic_score < 101')['user_score']
# средняя оценка по иксбоксу
print('Средняя оценка Xbox One =', for_xbx1.mean())
# средняя оценка по пк
print('Средняя оценка PC =', for_pc.mean())
```

```python
alpha = 0.05 # критический уровень статистической значимости, если pvalue больше него, то гипотеза будет отвергнута
results = stats.ttest_ind(for_xbx1, for_pc, equal_var=False) # дисперсии не равны поэтому False
print('pvalue = ', results.pvalue)

if (results.pvalue < alpha):
    print('Нулевую гипотезу получилось отвергнуть')
else:
    print('Нулевую гипотезу не получилось отвергнуть')
```

Тест показал, что:
- pvalue =  0.05,
- нулевая гипотеза подтверждена, следовательно нужно её принять - средние рейтинги юзеров Xbox One и PC равны.

### средние пользовательские рейтинги жанров Action и Sports разные

Два условия:

- H0 средние рейтинги юзеров Action и Sports одинаковые

- H1 средние рейтинги юзеров Action и Sports разные

Нужно сделать Series для игр с оценками пользователей по нужному срезу лет. Из стобца name мне нужны только action и sports, а также определённые выше года с 2006 по 2016. Так как мне нужно использовать - user_score.

```python
for_action = games.query('genre == "action" and year_of_release >= 2014 and user_score < 101 and critic_score < 101')['user_score']
for_sports = games.query('genre == "sports" and year_of_release >= 2014 and user_score < 101 and critic_score < 101')['user_score']
# средняя оценка по экшнам
print('Средняя оценка Action =', for_action.mean())
# средняя оценка по спортивным играм
print('Средняя оценка Sports =', for_sports.mean())
```

```python
alpha = 0.05 # критический уровень статистической значимости, если pvalue больше него, то гипотеза будет отвергнута
results = stats.ttest_ind(for_action, for_sports, equal_var=False)
print('pvalue=', results.pvalue)

if (results.pvalue < alpha):
    print('Нулевую гипотезу получилось отвергнуть')
else:
    print('Нулевую гипотезу не получилось отвергнуть')
```

Тест показал, что:
- pvalue = 6.9667,
- нулевую гипотезу получилось отвергнуть соответственно принимаю альтернативную - средние рейтинги юзеров Action и Sports разные.
